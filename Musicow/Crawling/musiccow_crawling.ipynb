```python
import requests
from bs4 import BeautifulSoup

song_list= list()

for j in range(1000):
    req = requests.get('https://www.musicow.com/song/'+str(j)+'?tab=info')
    html = req.text
    soup = BeautifulSoup(html,'html.parser')
    
    temp = list()

    title = soup.select(
        'strong.title'
    )
    
    artist = soup.select(
        'em.artist'
    )

    my_titles = soup.select(
        'span.bar'
    )
    recent_price = soup.select(
        'dl.price > dd.txt_red'
    )
    
    
    if title[0].text=='':
        continue
    #print("곡번호-"+str(j)+","+title[0].text+" by "+artist[0].text)
    temp.append(j)
    temp.append(title[0].text)
    temp.append(artist[0].text)
    #print("최근거래가 : "+recent_price[0].text.split(' ')[0])
    temp.append(int(recent_price[0].text.split(' ')[0].replace(',','')))
    for i in (1,2,3,4,5):
        #print(my_titles[i].get('data-bar-label')+"년 : "+ my_titles[i].get('data-bar-value')+"원")
        if my_titles[i].get('data-bar-value') == '':
            temp.append(0)
        else:
            temp.append(int(my_titles[i].get('data-bar-value').replace(',','')))
        #print()
    song_list.append(temp)
```
